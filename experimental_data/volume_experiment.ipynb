```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
volume = []
cap1 = []
cap2 = []

with open("capacitance_experiment_1.txt", 'r') as file:
    for line in file:
        line = line.split()

        volume.append(line[0])
        cap1.append(line[1])
        cap2.append(line[2])

x_labels = [str(i) for i in volume]
cap1_labels = [str(i) for i in cap1]
cap2_labels = [str(i) for i in cap2]
```
